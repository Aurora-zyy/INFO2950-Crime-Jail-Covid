```python
# Helper function generate regression outputs and plot
def model_output(X, Y, summary=False, plot=False, title="Linear Regression"):
    model = sm.OLS(Y, X).fit()
    predictions = model.predict(X)
    if plot:
        plt.scatter(pd.DataFrame(X).iloc[:,1], Y, color = 'blue')
        plt.plot(pd.DataFrame(X).iloc[:,1], predictions, color = 'red')
        plt.title(title)
        plt.xlabel('COVID predictors')
        plt.ylabel('Crime cases')
    if summary:
        return model.summary()
```

```python
# Bollinger Bands Helper Functions
def BollingerBands(data, col, days, noStd, method, plot=False):
    """
    Parameter data: the data containing column [col]
    Parameter days: window of moving average
    Parameter noStd: number of standard deviation, the width of bands
    Parameter method: takes either SMA(Simple) or EWMA(Exponentially Weighted)
    Return: updated dataframe with 
            [MA, MA STD, MA Upper Bound, MA Lower Bound, BreakoutSignal]
    """
    # Simple Moving Average
    if method == "SMA":
        data['MA'] = data[col].rolling(window=days).mean()
        data['MA_STD'] = data[col].rolling(window=days).std()
        data['MA_Upper'] = data['MA'] + data['MA_STD'] * noStd
        data['MA_Lower'] = data['MA'] - data['MA_STD'] * noStd
        
    # Exponentially Weighted Moving Average
    else:
        data['MA'] = data[col].ewm(span=days, adjust=False).mean()
        data['MA_STD'] = data[col].ewm(span=days, adjust=False).std()
        data['MA_Upper'] = data['MA'] + data['MA_STD'] * noStd
        data['MA_Lower'] = data['MA'] - data['MA_STD'] * noStd
    
    breakout_upper = np.where(data[col] > data['MA_Upper'], 1, 0)     # +1 when hit upper band
    breakout_lower = np.where(data[col] < data['MA_Lower'], -1, 0)    # -1 when hit lower band
    
    data['BreakoutSignal'] = breakout_upper + breakout_lower
    data['BreakoutSignal'].fillna(method='ffill',inplace=True)    # Forward fill to replace the "None" values
    
    # plot bands
    if plot:
        plt.plot( data[col], label="Training Data" )
        plt.plot( data[['MA']], alpha=0.7, label="Moving Average")
        plt.plot( data[['MA_Lower']], alpha=0.7, label="Lower Bound")
        plt.plot( data[['MA_Upper']], alpha=0.7, label="Upper Bound")
        plt.scatter(data[data["BreakoutSignal"] == -1].index, data[data["BreakoutSignal"] == -1][col], c='g', label="Lower Signal")
        plt.scatter(data[data["BreakoutSignal"] == 1].index, data[data["BreakoutSignal"] == 1][col], c='r', label="Upper Signal")
        plt.title( "{} days {} Bollinger Bands of {}".format(str(days), method, col) )
        plt.ylabel(col)
        plt.legend()
```

## 4.2 Bollinger Bands on COVID daily data
At this point, we have determined the relationship between the pandemic situation and the crime rate after selecting from a variety of regression methods. More importantly, we are interested in the future behavior of COVID. It’s undeniable that the pandemic influences the population socially and economically and studying its movement can help inform decisions. Before getting quantitative, we will first use Bollinger Bands to obtain some volatility analysis on the trend and fluctuation of COVID data.

**Data that we used:**
For both methods of forecasting, the Bollinger Bands Analysis and the following quantitative forecast, we will use `coviddata`, which contains daily covid and crime data for both cities. First, we divide the full data into training, validation and testing datasets. ### 4.2.1 NY Bollinger Bands Analysis
**Parameters of Bollinger Bands(BB)**
1. Window of Moving Average(MA): <br>
    For NY COVID data, we choose an optimal mid-value of a 15-day lookback window that is most sensitive to the training data.
2. Standard Deviation of BB: <br>
    It is common to use 2 standard deviations as the width of the bands.
3. Method of MA: <br>
    We attempted two types of MA: Simple Moving Average(SMA) and Exponentially Weighted Moving Average(EWMA). EWMA gives more weight to recent observations and less weight to observations further away from today. Though it sounds more intuitive than SMA, EWMA produces loose bands that cause insensitivity for high volatility data. Therefore, we will use SMA for later analysis. 


See below for the Bollinger Bands plots for NY Cases and Deaths. 

```python
plt.subplots(1,2, figsize=(20,5))
plt.subplot(1, 2, 1)
BollingerBands(coviddata, 'NYCase', 15, 2, "SMA", True)
plt.axvline('2020-09-29', -2000, 7000, color="gray", linestyle='dashed')
plt.plot(covidvalid['NYCase'], label="Validation Data", color='Black')
plt.axvline('2020-11-19', -2000, 7000, color="gray", linestyle='dashed')
plt.plot(covidtest['NYCase'], label="Testing Data", color='purple')
plt.legend()

plt.subplot(1, 2, 2)
BollingerBands(coviddata, 'NYDeath', 15, 2, "SMA", True)
plt.axvline('2020-09-29', -2000, 7000, color="gray", linestyle='dashed')
plt.plot(covidvalid['NYDeath'], label="Validation Data", color='Black')
plt.axvline('2020-11-19', -2000, 7000, color="gray", linestyle='dashed')
plt.plot(covidtest['NYDeath'], label="Testing Data", color='purple')
plt.legend()
plt.show()
```

The `blue curve`, `black curve` and `purple curve` are the training, validation, and testing daily cases and deaths of NY. The `yellow curve` is the 15-day moving average of daily cases and deaths. The `red curve` and `green curve` are the upper and lower Bollinger Band, which are 2 standard deviations from the moving average. The `red dots` and `green dots` indicate observed counts exceeding the upper and lower limit. 


**NY COVID Training Data Analysis: left side of dashed line**<br>
>**Cases (on the left)**
>- Looking to the very left, we observe that the early phase of the pandemic has extremely large volatility during March and April. The Bollinger Bands successfully detected the increasing trend of covid cases in March. With the consecutive red dots, the number of cases breaks out of the upper bound. It is not until the end of April, that the bands give a decreased signal with two green dots. 
>- In September, there are three red upper bound breakout signals and the bands begin to expand. **We expect the number of cases to increase in the following months, but only by a small amount.**


>**Deaths (on the right)**
>- We observe very similar signals for case and death, as expected. The trend of death cases slightly lags behind positive cases. Very quickly in mid-April, as soon as cases start to drop, we see green dots, indicating more decrease in deaths. 
>- Up to October, we observe no breakout signals, so **we expect the number of deaths to reduce and stay low in the following months.**

## 4.3 Forecasting COVID daily data with ARIMA models
Now that we have an idea of current and future COVID cases and deaths of both cities. How can we go further and make quantitative COVID predictions?

We will use a model called the autoregressive integrated moving average (ARIMA) model. This model and the exponential smoothing model are the two most widely used approaches to time series forecasting. Since the COVID data is essentially a time series. Also, research suggests that the ARIMA model performs better on COVID data than the exponential smoothing model. So, we chose ARIMA as our forecasting model (Papastefanopoulos et al., 2020).

The modeling method we used here is the auto_arima() function in the pmdarima package. This function can automatically make adjustments on parameters and evaluate multiple ARIMA models' performances to choose the best one with the lowest residuals. There are 7 parameters in an ARIMA model. If you are not familiar with ARIMA and related concepts, here is a nice tutorial: https://otexts.com/fpp2/arima.html

In this section, we will build ARIMA models on COVID daily cases and deaths of NY and LA. See section 4 in `Complete_Data_Analysis.ipynb` for omitted training and validation steps for the model we will use below. In the validation process, our model provides best accuracy within the first two weeks of forecast. This is not a problem because, in reality, the model should be updated with new daily observations and make short-term predictions to ensure accuracy.


### 4.3.1 NY ARIMA Analysis
Before calling the `auto_arima()` function, we will perform some exploratory study of data seasonality in order to help determine specific values used for parameters. The left column has diagnostic plots for daily cases, and the `right column` has diagnostic plots for daily deaths. 


We will examine an `autocorrelation plot (ACF)` of the data. The ACF plot presents the correlation between the data on one day and 'n' day before. Here, the number n is called ‘lag’; it is presented on the x-axis of the plots below. The y-value is the correlation coefficient between those two points. The plot tells us the effects one day's data acts on the days following, or you can take it as the similarity between those two days. The higher the y-value, the more similar those two points are. These two plots play a key role in determining the seasonality parameter for the `auto_arima()` function.

```python
# make ACF plots for modeling purposes.
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
plot_acf(coviddata["NYCase"], ax=axs[0])
plot_acf(coviddata["NYDeath"], ax=axs[1])
plt.show()
```

>From the ACF plot on the left, we observe an apex every seven-days, which means that the NYCase data follows a seven-day periodicity. The ACF plot of NYDeath on the right, on the other hand, shows little periodicity suggesting very little seasonality in the NYDeath data. We will build our ARIMA models accordingly.



Next, we can go ahead with our ARIMA model. The process of parameter selection involves:
>The function generates a different version of the ARIMA models and chooses AIC as the estimator of prediction error. After looking through possible candidates of each parameter, the function returns an ARIMA model with the lowest AIC.


**For readability consideration, we will omit the tuning process of the all ARIMA models.**

See above for the ARIMA plots for NY Cases and Deaths. The `blue curve` is the observed actual daily cases and deaths of NY. The `orange curve` is the two-week forecast data of daily cases and deaths.

```python
# NY Cases
# Develop an ARIMA model using auto ARIMA function and predict 14 days
forecast_arima_model_ny_case = auto_arima(coviddata["NYCase"], start_p=0, error_action='ignore', m=7, seasonal=True, D = 2,suppress_warnings=True, stepwise=True, random_state=20, n_fits=50)
forecast_prediction_ny_case = pd.DataFrame(forecast_arima_model_ny_case.predict(n_periods = 14), index=pd.date_range('2020-12-03', periods=14, freq='D'))
forecast_prediction_ny_case.columns = ['predicted_cases_ny']

# NY Deaths
# Develop an ARIMA model using auto ARIMA function and predict 14 days
forecast_arima_model_ny_death = auto_arima(coviddata["NYDeath"], start_p=0, error_action='ignore',suppress_warnings=True, stepwise=True, random_state=20, n_fits=50)
forecast_prediction_ny_death = pd.DataFrame(forecast_arima_model_ny_death.predict(n_periods = 14), index=pd.date_range('2020-12-03', periods=14, freq='D'))
forecast_prediction_ny_death.columns = ['predicted_deaths_ny']
```

```python
# visualize the prediction
plt.subplots(1,2, figsize=(15,5))
plt.subplot(1, 2, 1)
plt.plot(coviddata["NYCase"][-30:], label='Train')
plt.plot(forecast_prediction_ny_case['predicted_cases_ny'], label='Prediction')
plt.xlabel('Date')
plt.xticks(rotation=15)
plt.ylabel('Number of Cases')
plt.legend(prop={'size': 12})
plt.title('The Line Plot of NYCase and its Prediction')

plt.subplot(1, 2, 2)
plt.plot(coviddata["NYDeath"][-30:], label='Train')
plt.plot(forecast_prediction_ny_death['predicted_deaths_ny'], label='Prediction')
plt.xlabel('Date')
plt.xticks(rotation=15)
plt.ylabel('Number of Death')
plt.legend(prop={'size': 12})
plt.title('The Line Plot of NYDeath and its Prediction')
plt.show()
```

**NY ARIMA Evaluation with Testing Data**
>- For NY cases (left), we observe the same growing trend with greater volatility in the next 2 weeks. On average the positive cases will rise from around 1700 to 2500 per day. For NY deaths (right), we expect the number of daily deaths to increase from around 25 to 40 by mid-December.
>- Both predictions match the previous Bollinger Band's expectations of worsening pandemic conditions in December. We have numerical evidence showing the climb of a new wave.

# 5. Data limitations #

- Assume the reliability/authenticity of data sources.
- Assume zero covid cases before March 2020 due to missing data records.
- The latest data used is up-to-date as of November 2020, so the results might change with additional data. 
- Our study only applies to the two metropolitan cities of focus, New York City, and Los Angeles. The results of our findings should not be generalized to other cities or suburban/rural regions in the US.
- Due to the lack of city-level COVID data for Los Angeles, we are replacing it with county-level data to assist our analysis. However, the county-level data might be more generalized and cause inconsistencies with other LA datasets. 
- Lack of daily economic data of various industries. The intensity of the COVID pandemic could change dramatically within a month. The monthly statistics cannot reveal timely or accurately the correlation between the industries' variables and pandemic situations. Since the pandemic lasts for less than a year, the size of the monthly dataset is largely limited. It is almost impossible to generate robust regression analysis or any other quantitative analysis. 
- All employment data are not seasonally adjusted. The real pattern could possibly be disturbed by the seasonal noises.
- Due to the availability of daily data, the study focuses solely on the relationship between COVID and crime data. Such relation is subject to any additional variable effects on crime rates.
- For any method of time series, the prediction power decreases when forecasting dates that are far into the future. Our prediction of COVID cases is most accurate for the recent future. Ideally, one should update forecast algorithms daily, with up-to-date data.
